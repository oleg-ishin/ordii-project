# Загрузка и первичный обзор

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("google_books_dataset.csv")
```

```python
print("Первые 5 строк: ")
df.head(5)
```

```python
print("Последние 5 строк: ")
df.tail(5)
```

```python
print('Структура данных датасета:')
print(df.info())


df['published_date'] = pd.to_datetime(df['published_date'], errors='coerce')
df['publication_year'] = df['published_date'].dt.year
df['isbn_13'] = df['isbn_13'].astype('Int64').astype(str)
```

# Обработка пропусков и дубликатов

```python
print('Пропуски:')
print(df.isna().sum())

numeric_cols = ['published_date', 'page_count', 'average_rating', 'ratings_count', 'list_price']
str_cols = ['book_id', 'title','subtitle','authors','publisher','description','categories','language','preview_link','info_link','isbn_13','isbn_10','currency','buyable','search_category','thumbnail']

df.dropna()

for i in numeric_cols:
  df[i] = df[i].fillna(0)

for i in str_cols:
  df[i] = df[i].fillna('Unknown')

print("Количество дубликатов до удаления:    %3d" % df.duplicated().sum())
df.drop_duplicates()
print("Количество дубликатов после удаления: %3d" % df.duplicated().sum())
```

# Статистики данных

```python
print('Статистика числовых значений датасета:')
print(df.describe())
```

```python
print('Количество уникальных значений:')
print(df.nunique())
```

```python
print('Таблица корреляций:')
df.corr(numeric_only=True)
```

# Группировки и агрегации

```python
print("Топ 5 популярных языков")
books_by_lang = df.groupby('language').size().sort_values(ascending=False)
print(books_by_lang.head(5))

print("Среднее количество страниц по издателям:")
mean_pages_by_publisher = df[df['publisher'] != 'Unknown'].groupby('publisher')['page_count'].mean().sort_values(ascending=False)
print(mean_pages_by_publisher.head(5))

print("Количество отзывов в каждой категории:")
total_ratings_by_category = df[df['categories'] != 'Unknown'].groupby('categories')['ratings_count'].sum().sort_values(ascending=False)
print(total_ratings_by_category.head(5))

print("\nСводная таблица:")
pivot_table_data = df.pivot_table(
    index='publication_year',
    columns='language',
    values='book_id',
    aggfunc='count'
)
print(pivot_table_data.tail(10))
```
